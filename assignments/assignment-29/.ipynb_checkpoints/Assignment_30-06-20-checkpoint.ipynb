```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/wbcd.csv")
df.head()
```

```python
df.drop("id",axis=1,inplace=True)
df.head()
```

```python
df[df.duplicated()]
```

```python
df.nunique()
```

```python
di=pd.get_dummies(df.diagnosis,drop_first=True)
di
```

```python
vis=df.drop("diagnosis",axis=1)
for i in vis.columns:
    plt.xlabel("diagnosis")
    plt.ylabel(i)
    plt.scatter(di
                ,vis[i])
    plt.show()
```

```python
df.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr()[df.corr()>0.6],annot=True)
```

```python
sns.countplot(df.diagnosis)
```

```python
x=df.drop("diagnosis",axis=1)
x
```

```python
y=df.diagnosis
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
scores=[]
for i in range(1,20):
    knn=KNeighborsClassifier(n_neighbors=i,metric="euclidean")
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    scores.append(accuracy_score(y_test,y_pred))
scores
```

```python
plt.plot(scores,'k--',marker="o",markerfacecolor="red")
```

```python
print(scores[8],scores[14])
```

When n_neighbor=9 and 15 Shows a high accuracy

```python
final_model=KNeighborsClassifier(n_neighbors=9,metric="euclidean")
final_model.fit(x_train,y_train)
final_y_pred=final_model.predict(x_test)
accuracy_score(y_test,final_y_pred)
```

```python
confusion_matrix(y_test,final_y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,final_y_pred),annot=True,annot_kws={'size':20})
```

```python
print(classification_report(y_test,final_y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

sc_x_train=pd.DataFrame(sc.fit_transform(x_train),columns=x_train.columns)
sc_x_train.head()
```

```python
sc_x_test=pd.DataFrame(sc.transform(x_test),columns=x_test.columns)
sc_x_test.head()
```

```python
sc_scores=[]
for i in range(1,20):
    knn=KNeighborsClassifier(n_neighbors=i,metric="euclidean")
    knn.fit(sc_x_train,y_train)
    sc_y_pred=knn.predict(sc_x_test)
    sc_scores.append(accuracy_score(y_test,sc_y_pred))
sc_scores
```

```python
plt.plot(sc_scores,'k--',marker="o",markerfacecolor="red")
```

```python
sc_final_model=KNeighborsClassifier(n_neighbors=7,metric="euclidean")
sc_final_model.fit(sc_x_train,y_train)
sc_final_y_pred=sc_final_model.predict(sc_x_test)
accuracy_score(y_test,sc_final_y_pred)
```

```python
confusion_matrix(y_test,sc_final_y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,sc_final_y_pred),annot=True,annot_kws={'size':20})
```

```python
print(classification_report(y_test,sc_final_y_pred))
```

Scalared Data has more accuracy than Original Data
